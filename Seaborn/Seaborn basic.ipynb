```python
iris = sns.load_dataset("iris")
```

```python
sns.relplot(x= iris.petal_width, y =iris.petal_length,data = iris,hue= 'species') # jya 3 categories che species ne ena 3 group na alag alag color ma represent kare che
```

```python
sns.relplot(x=iris.petal_width, y = iris.petal_length , data = iris ,hue= 'petal_width') # hue e color aape che je data available hoi ne ene 
```

```python
sns.relplot(x=iris.petal_width , y = iris.petal_length,data= iris,style = 'species',hue='petal_length')
```
