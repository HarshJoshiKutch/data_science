```python
data1 = pd.DataFrame({'key1':[1,2,3,4,5],
                      'key2':[6,7,8,9,10],
                      'key3':[3,6,4,2,9]
}
)
```

```python
data2 = pd.DataFrame({'key1':[1,2,3,4,5],
                      'key4':[22,7,8,49,10],
                      'key5':[3,65,42,21,9]
}
)
```

```python
data2
```

have aama aapne merge function no use karsu, merge e teni column value par kaam kare che


```python
pd.merge(data1,data2) # ama je common data hase e lese teni under 
```

```python
pd.merge(data1,data2,how = 'left') # aama je left ma etle je data1 hase e and right ma data 2 aavse
```

```python
pd.merge(data1,data2,how = 'right') # aama je left ma etle je data1 hase e and right ma data 2 
```

```python
pd.merge(data1,data2,how ='cross') # cartesian product karine bhadha ne ek bija sathe match karve che
```
