```python
x1 = np.linspace(0,10,50)
y1 = np.linspace(0,10,50)
plt.scatter(x1,y1)
```

```python
colours = np.random.rand(50) # aani help thi colours pan decide kari sakay che 
sizes = 1000*np.random.rand(50) # sizes aapi sakiye chiye data ne 
plt.scatter(x1,y1,c = colours , s = sizes , alpha = .8)# alpha te intensity ne represent kare che jya .1to maja ave tela rakhi sakay
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("random data")
```

```python
x = np.linspace(0,10,200)
y = np.sin(x)
```

```python
plt.figure(figsize=(8,4))
plt.plot(x,y,'--b') # ther is b is denoted for the blue color you can also give red color for r there 
plt.xlabel("This is my x data")
plt.ylabel("This is my y data")
plt.title("This is my title")
plt.show()
```

```python
plt.figure(figsize=(6,2)) # jya tenu size reduce kari de che 
plt.plot(x,y,'--b')  
plt.xlabel("This is my x data")
plt.ylabel("This is my y data")
plt.title("This is my title")
plt.grid() # jya pachad grid means lines draw thay che 
plt.show()
```
